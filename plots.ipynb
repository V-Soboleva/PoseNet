```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
model = 'KingsCollege'



pose_true = pd.read_csv('pose_true.csv')
pose_estim = pd.read_csv('pose_estim.csv')
test_res = pd.read_csv('test_result.csv')

n = 10

position_true = pose_true.iloc[n, 0:3].values
position_estim = pose_estim.iloc[n, 0:3].values


fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# r = [-1, 1]
# for s, e in combinations(np.array(list(product(r, r, r))), 2):
#     if np.sum(np.abs(s-e)) == r[1]-r[0]:
#         ax.plot3D(*zip(s, e), color="b")

r = np.sqrt(test_res.iloc[n, 2])
u, v = np.mgrid[0:2*np.pi:20j, 0:2*np.pi:20j]
x = r * np.cos(u)*np.sin(v) + position_estim[0]
y = r * np.sin(u)*np.sin(v) + position_estim[1]
z = r * np.cos(v) + position_estim[2]

ax.plot_wireframe(x,y,z,color = 'b', alpha = 0.1)
ax.scatter(position_true[0], position_true[1], position_true[2], c='r', marker='o', label = 'turth')
ax.scatter(position_estim[0], position_estim[1], position_estim[2], c='b', marker='o', label = 'estimation')


ax.legend()

# plt.axis('scaled')
plt.show();
```
